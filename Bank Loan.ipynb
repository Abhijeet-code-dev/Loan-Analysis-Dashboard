DASHBOARD QUESTION SOLN'S 

## FIND PRIMARY KPI'S 

```python
#1. Total Loan applications
total_loan_applications = df['id'].count()
print(f"Total Loan applications: {total_loan_applications}")

#2. Month to Date (MTD) Total Loan Applications
#Month to Date means current month ka data For eg: this data has data till december so we have to find total loan applications for dec
#In realtime, current month for which u r doing analysis is month to date data of that yr 

latest_issue_date = df['issue_date'].max()
latest_year = latest_issue_date.year
latest_month = latest_issue_date.month
#print(latest_month)

#Matching yr bcz agar suppose 5 yrs ka data ho 
mtd_data = df[(df['issue_date'].dt.year == latest_year) & (df['issue_date'].dt.month == latest_month)]
mtd_loan_applications = mtd_data['id'].count()
print(f"Total MTD Loan applications: {mtd_loan_applications} (for {latest_issue_date.strftime('%B %Y')})")
```

```python
#3. Total funded amount means total loan amount
total_funded_amount = df['loan_amount'].sum()
print(f"Total Funded Amount is: {total_funded_amount}")

#Total funded amount in million - to get this divide by 10l
total_funded_amt_millions = total_funded_amount/1000000

print("Total Funded Amount: ${:.2f}M".format(total_funded_amt_millions))
#Syntax explanations $ - for currency symbol , {:.2f} - means 2 digits after decimal , ${:.2f}M - means aage $ aye aur piche M for Millions
```

```python
#4. MTD - Total Funded Amount
latest_issue_date = df['issue_date'].max()
latest_issue_year = latest_issue_date.year
latest_issue_month = latest_issue_date.month
mtd_data = df[(df['issue_date'].dt.year == latest_issue_year) & (df['issue_date'].dt.month == latest_issue_month)]
total_loan_amt = mtd_data.loan_amount.sum()
total_loan_amt_millions = total_loan_amt/1000000
print("Total MTD Loan Amount: ${:.2f}M".format(total_loan_amt_millions))
```

```python
#5. Total amt received ie. loan repaid
total_amt_received = df['total_payment'].sum()
total_amt_received_millions = total_amt_received/1000000
print("Total amt received ${:.2f}M".format(total_amt_received_millions))

#MTD total amt received
latest_issue_date = df['issue_date'].max()
latest_issue_year = latest_issue_date.year
latest_issue_month = latest_issue_date.month
mtd_data = df[(df['issue_date'].dt.year == latest_issue_year) & (df['issue_date'].dt.month == latest_issue_month)]
total_received_amt = mtd_data.total_payment.sum()
total_received_amt_millions = total_received_amt/1000000
print("Total MTD Received Amount: ${:.2f}M".format(total_received_amt_millions))
```

Good Loan Metrics

```python
#creating df containing good loans data
good_loans = df[df['loan_status'].isin(['Current','Fully Paid'])]
#total good loans
good_loan_applications = good_loans['id'].count()
print("Good Loan applications are:",good_loan_applications)
total_loan_applications = df['id'].count()

#Good Loans funded amount
good_loans_funded_amt = good_loans['loan_amount'].sum()
loans_funded_amt_millions = good_loans_funded_amt/1000000
print("Good Loans Funded Amount: ${:.2f}M".format(loans_funded_amt_millions))

#Good Loans total received amt
good_loans_received_amt = good_loans['total_payment'].sum()
loans_received_amt_millions = good_loans_received_amt/1000000
print("Good Loans Received Amount: ${:.2f}M".format(loans_received_amt_millions))

#Good Loan application %age
loan_application_page = good_loan_applications*100/total_loan_applications
print("Good Loan application percentage:{:.2f}%".format(loan_application_page))
```

Bad Loan Metrics

```python
#creating df containing bad loans data
bad_loans = df[df['loan_status'].isin(['Charged Off'])]
#total bad loans
bad_loan_applications = bad_loans['id'].count()
print("Bad Loan applications are:",bad_loan_applications)
total_loan_applications = df['id'].count()

#bad Loans funded amount
bad_loans_funded_amt = bad_loans['loan_amount'].sum()
loans_funded_amt_millions = good_loans_funded_amt/1000000
print("Bad Loans Funded Amount: ${:.2f}M".format(loans_funded_amt_millions))

#bad Loans total received amt
bad_loans_received_amt = bad_loans['total_payment'].sum()
loans_received_amt_millions = bad_loans_received_amt/1000000
print("Bad Loans Received Amount: ${:.2f}M".format(loans_received_amt_millions))

#bad Loan application %age
loan_application_page = bad_loan_applications*100/total_loan_applications
print("Bad Loan application percentage:{:.2f}%".format(loan_application_page))
```

Monthly trends by issue date for total amount received

```python

#creating data
monthly_recieved = (
    df.sort_values('issue_date') #Sorting by month
    .assign(month_name = lambda x: x['issue_date'].dt.strftime('%b %Y')) #%b- shorten month names , creating month_name new col here
    .groupby('month_name',sort=False)['loan_amount']
    .sum()
    .div(1000000)
    .reset_index(name = 'total_recieved')
)
plt.figure(figsize=(10,5))

x = np.arange(len(monthly_recieved))  # use monthly_recieved
y = monthly_recieved['total_recieved'].to_numpy()  # use monthly_recieved

plt.fill_between(x, y, color='lightgreen', alpha=0.5)
plt.plot(x, y, color='green', linewidth=2)

for i, val in enumerate(y):
    plt.text(i, val + 0.1, f"{val:.2f}", ha='center', va='bottom', fontsize=9, color='black')

plt.title('Total Payments Received by Month', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Payment Amount Received ($ Millions)')
plt.xticks(ticks=x, labels=monthly_recieved['month_name'], rotation=45)

plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

```

Monthly trends by issue date for total Loan applications

```python


# ----------------- Prepare Data -----------------
monthly_applications = (
    df.sort_values('issue_date')  # Sorting by month
      .assign(month_name=lambda x: x['issue_date'].dt.strftime('%b %Y'))  # Short month names
      .groupby('month_name', sort=False)['id']
      .count()  # Count of applications
      .reset_index(name='loan_applications_count')  # Reset column name
)

# ----------------- Plot -----------------
x = np.arange(len(monthly_applications))
y = monthly_applications['loan_applications_count'].to_numpy()

plt.figure(figsize=(10,5))
plt.fill_between(x, y, color='orange', alpha=0.5)
plt.plot(x, y, color='orange', linewidth=2, marker='o')

# Add value labels on top of points
for i, val in enumerate(y):
    plt.text(i, val + 2, f"{val}", ha='center', va='bottom', fontsize=9, color='black')

plt.title('Total Loan Applications Month-wise', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Total Loan Applications')
plt.xticks(ticks=x, labels=monthly_applications['month_name'], rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

Regional analysis by state for total amount funded

```python

state_funded = (
    df.groupby('address_state')['loan_amount']  
      .sum()
      .sort_values(ascending=True) 
      .reset_index()
)

plt.figure(figsize=(10,12))
bars = plt.barh(state_funded['address_state'], state_funded['loan_amount'], color='skyblue')

plt.xlabel("Total Funded Amount")
plt.ylabel("State")
plt.title("Regional Analysis: Total Funded Amount by State")

for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.01*width, bar.get_y() + bar.get_height()/2, f"{int(width):,}", 
             va='center', fontsize=9, color='black')

plt.tight_layout()
plt.show()
```

Long Term Analysis by total funded amount

```python
term_summary = df.groupby('term')['loan_amount'].sum().reset_index()

term_summary['percentage'] = (term_summary['loan_amount'] / term_summary['loan_amount'].sum()) * 100

plt.figure(figsize=(7,7))
colors = ['skyblue', 'lightgreen']
plt.pie(term_summary['loan_amount'], 
        labels=term_summary['term'] + ' (' + term_summary['percentage'].round(1).astype(str) + '%)',
        colors=colors,
        autopct='%1.1f%%',
        startangle=140,
        explode=[0.05]*len(term_summary)
       )
plt.title('Long Term Loan Analysis by Total Funded Amount')
plt.show()
```
